```python
import getopt
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.ticker as mticker
```

```python
data = pd.read_csv('/Users/winnymaisyarah/Documents/digital skola/k-means-cluster/input_data/Mall_Customers.csv')

print(data)
```

```python
x = data.iloc[1::, [3, 4]].values

print(x)
```

```python
n_cluster = 5
kmeans = KMeans(n_cluster, init = 'k-means++')
kmeans.fit_predict(x)
```

```python
identified_clusters = kmeans.labels_

print(identified_clusters)
```

```python
data_with_clusters = data.copy()

data_with_clusters
```

```python
fig = plt.figure(figsize=(12,10))
plt.scatter(x[identified_clusters == 0, 0], x[identified_clusters == 0, 1], s = 100,
            c = "green", label = "High Spend High Income")
plt.scatter(x[identified_clusters == 1, 0], x[identified_clusters == 1, 1], s = 100,
            c = "red", label = "Low Spend High Income")
plt.scatter(x[identified_clusters == 2, 0], x[identified_clusters == 2, 1], s = 100,
            c = "purple", label = "Medium Spend Medium Income")
plt.scatter(x[identified_clusters == 3, 0], x[identified_clusters == 3, 1], s = 100,
            c = "yellow", label = "Low Spend Low Income")
plt.scatter(x[identified_clusters == 4, 0], x[identified_clusters == 4, 1], s = 100,
            c = "blue", label = "High Spend Low Income")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            s = 150, c = "black", label = "Centroids")
plt.title('Clustered Mall Using K-Means')
plt.xlabel('Income (Juta Rupiah)')
plt.ylabel('Spending (Juta Rupiah)')

plt.gca().xaxis.set_major_formatter(mticker.FixedFormatter(['{:}'.format(x/1e6) for x in plt.gca().get_xticks()]))
plt.gca().yaxis.set_major_formatter(mticker.FixedFormatter(['{:}'.format(y/1e6) for y in plt.gca().get_yticks()]))
plt.gca().xaxis.set_major_locator(mticker.FixedLocator(plt.gca().get_xticks()))
plt.gca().yaxis.set_major_locator(mticker.FixedLocator(plt.gca().get_yticks()))
plt.legend()
plt.savefig("/Users/winnymaisyarah/Documents/digital skola/k-means-cluster/output_data/Clustered-Mall-Using-K-Means.png")
plt.show()
```
